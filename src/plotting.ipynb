```python
#plot runtime dataset
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("../results/scalability_1d.csv")
df.head()

# mode,bias_kind,nbins,domain_size,label,groups,time_build,time_star,util_count,util_theta
# full,per_value,,65,baseline,1,5.908403545618057e-05,0.0,40,17.939393939393934
# full,per_value,,65,q_base,1,9.664721500012092,0.0,40,17.939393939393934
# full,per_value,,65,q_star,1,9.664721500012092,9.849993512034416e-05,40,17.939393939393934

#plot time_build vs domain_size for label baseline, q_base, q_star
for label in df['label'].unique():
    sub = df[df['label'] == label]
    plt.plot(sub['domain_size'], sub['time_build'], label=label)
plt.xlabel('Domain Size')
plt.ylabel('Time to build (s)')
plt.title('Time to build vs Domain Size')
plt.legend()
plt.show()
```
